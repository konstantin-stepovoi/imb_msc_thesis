```python
import numpy as np
import scipy.stats as stats

def generate_lognormal_value(median: float, lower_bound: float, upper_bound: float) -> float:
    # Находим параметры логнормального распределения
    mu = np.log(median)
    
    # Используем обратное нормальное распределение для нахождения z-значений
    z_lower = (np.log(lower_bound) - mu)  # Логарифмируем нижнюю границу
    z_upper = (np.log(upper_bound) - mu)  # Логарифмируем верхнюю границу
    
    # Нахождение стандартного отклонения
    sigma = (z_upper - z_lower) / (2 * 3.919927969080108)  # 0.975 для 95% доверительного интервала
    
    # Генерация случайного значения из логнормального распределения
    return np.random.lognormal(mean=mu, sigma=sigma)

# Пример использования функции
random_value = generate_lognormal_value(median=110000, lower_bound=10000, upper_bound=1000000)
print(f'Сгенерированное значение: {random_value}')
```
