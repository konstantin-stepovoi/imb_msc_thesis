```python
import os
import numpy as np
import matplotlib.pyplot as plt

#   gapsizes
gapsizes = [5000, 10000, 50000, 100000, 250000]
filenames = ['Rotatest_5000.txt', 'Rotatest_10000.txt', 'Rotatest_50000.txt', 'Rotatest_100000.txt']

#     
fig, axs = plt.subplots(2, 3, figsize=(15, 9))  # 3x3  9 
fig.suptitle("    ", fontsize=16)

#      
for i in range(4):
    #   
    filename = filenames[i]
    
    #     
    with open(filename, 'r') as file:
        lines = file.readlines()[1:]  #   
        data = np.array([float(line.strip()) for line in lines])  #    float

    #      
    mean_value = np.mean(data)
    std_dev = np.std(data)

    #    
    row, col = divmod(i, 3)  #     3x3
    ax = axs[row, col]

    #  
    ax.hist(data, color='skyblue', edgecolor='black')
    ax.set_title(f'{filenames[i]} \nMean: {mean_value:.2f}, Std: {std_dev:.2f}')
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.set_xlim(mean_value - 3 * std_dev, mean_value + 3 * std_dev)


#    
plt.tight_layout(rect=[0, 0, 1, 0.96])  #     
plt.show()
```
